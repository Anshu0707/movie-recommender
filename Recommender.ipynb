```python
import pandas as pd
import numpy as np

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

Anshu_df = pd.read_csv("Anshu-csv.csv")
Anshu_df.head()
```

```python
def movieType(row):
    return row['genres']+row['tag']

Anshu_df["movieType"] = Anshu_df.apply(movieType, axis =1)
Anshu_df.head()
```

```python
vector = CountVectorizer()

countMatrix = vector.fit(Anshu_df["movieType"])
countMatrix = vector.transform(Anshu_df["movieType"])

print("Count Matrix:")
print()
print(countMatrix.toarray())
```

```python
dotProduct = cosine_similarity(countMatrix)
print(dotProduct)
```

```python
input_Movie = "Toy Story (1995)"

def index(title):
    return Anshu_df[Anshu_df.title == title]["index"].values[0]

movie_index = index(input_Movie)
print(movie_index)
```

```python
similar = list(enumerate(dotProduct[movie_index]))
print(similar)
```

```python
RecomList = sorted(similar, key=lambda x:x[1], reverse=True)
```

```python
def title(index):
    return Anshu_df[Anshu_df.index == index]["title"].values[0]

i=0
for movie in RecomList:
    print(title(movie[0]))
    i=i+1
    if i>20:
        break
```
